## сборка CT, RGB, labels в 1 файл

````python
rgb = np.ones((height, width, id_diff, 3), dtype=np.uint8)

for i, id in enumerate(tqdm(range(id_min, id_max + 1))):
    img = Image.open(path_to_data + f'rgb/rgb{id}.tif')
    rgb[:, :, i, :] = np.array(img)

print(rgb.shape)
````

````python
ct = np.ones((height, width, id_diff), dtype=np.uint8)

for i, id in enumerate(tqdm(range(id_min, id_max + 1))):
    img = Image.open(path_to_data + f'CT/frozenCT{id}.tif')
    ct[:, :, i] = np.array(img)

print(ct.shape)
````

````python
rgb = (rgb / 256).astype(np.float32)
ct = (ct / 256).astype(np.float32)
````

````python
np.save('data/numpy/rgb_unmasked.npy', np.ascontiguousarray(rgb))
np.save('data/numpy/ct_unmasked.npy', np.ascontiguousarray(ct))
np.save('data/numpy/labels.npy', np.ascontiguousarray(labels))
````
